```python
import random

# Define valid choices for the game
CHOICES = ["rock", "paper", "scissors"]



def get_player_choice():
    """Get user input and validate it is one of the choices above"""
    player_choice = None
    while player_choice is None:
        # Prompt the user for their choice and validate it against the valid choices
        player_choice = input('Choices: \nRock \nPaper \nScissors \n\nPick: ')
        if player_choice.lower() not in CHOICES:
            player_choice = None
    return player_choice.lower()

def get_computer_choice():
    """Have the computer pick one of the valid choices at random"""
    # Use random.choice to pick a choice from the list of valid options
    computer_choice = random.choice(CHOICES)
    return computer_choice

def is_draw(player_choice, computer_choice):
    """Check if game was a draw"""
    return player_choice == computer_choice

def print_winner(player_choice, computer_choice, player_choice_points, computer_choice_points):
    """Check to see who won and update points"""
    if player_choice == 'rock' and computer_choice == 'scissors':
        print('Player wins!')
        player_choice_points += 1
        print("Human has won")
        print("You have " + str(player_choice_points))
        
    elif player_choice == 'scissors' and computer_choice == 'paper':
        print('Player wins!')
        player_choice_points += 1
        print("Human has won")
        print("You have " + str(player_choice_points))
        
    elif player_choice == 'paper' and computer_choice == 'rock':
        print('Player wins!')
        player_choice_points += 1
        print("Human has won")
        print("You have " + str(player_choice_points))
        
    else:
        print('Computer wins!')
        print('%s beats %s' % (computer_choice, player_choice))
        computer_choice_points += 1
        print("Computer has " + str(computer_choice_points))
    
    return player_choice_points, computer_choice_points

def play_game():
    """Play the game"""
    player_choice_points = 0
    computer_choice_points = 0
    
    while True:
        # Get player and computer choices
        player_choice = get_player_choice()
        computer_choice = get_computer_choice()
        
        if is_draw(player_choice, computer_choice):
            print("It's a draw, both players picked %s: " % player_choice)
        else:
            print("Computer picked: %s" % computer_choice)
            print("Player picked: %s" % player_choice)
            # Determine the winner and update points
            player_choice_points, computer_choice_points = print_winner(player_choice, computer_choice, player_choice_points, computer_choice_points)
        
        # Display the current score
        print("Current Score - Player: %d, Computer: %d" % (player_choice_points, computer_choice_points))
        
        play_again = input("Do you want to play again? (yes/no): ")
        if play_again.lower() != "yes":
            break

if __name__ == "__main__":
    play_game()
```
